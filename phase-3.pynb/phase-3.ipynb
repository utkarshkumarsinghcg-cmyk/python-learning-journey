### Advance Operation With Business example

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("====Zomato sale analysis====")
print("\n sale data shape", sales_data.shape)
print("\n Sample data for 1st 3 result :", sales_data[  : ,:3])
```

```python
# Total sale per year

print(np.sum(sales_data, axis=0))

Yearly_sale = np.sum(sales_data[:,1:],axis=0)
print(Yearly_sale)


# Total sale

print(np.sum(Yearly_sale,axis=0))

# Minimum sale per restaurant

min_sale = np.min(sales_data[:,1:],axis=1) # use 1 to work on the data from left to right (rows)
print(min_sale)

#maximum sale per year

max_sale = np.max(sales_data[ : ,1:],axis=0)  # use 0 to work on the data from top to bottom (coloumn)
print(max_sale)


# Averege sale per restautant

avg_sale = np.mean(sales_data[:,1:],axis=1)
print(avg_sale)


# Cumilative sum

CumSum = np.cumsum(sales_data[:,1:],axis=1)
print(CumSum)

# matplotlib

plt.figure(figsize=(8,6))
plt.plot(np.mean(CumSum , axis=0))
plt.title("Average Cumulative sale accross all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
