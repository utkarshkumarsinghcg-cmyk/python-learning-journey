### List vs numpy array

```python
import numpy as np

py_list = [1,2,3]


print("python list multiplication",py_list*2)


np_array  = np.array([1,2,3]) #element wise multiplication
print("pyhon array multiplication",np_array*2)

import time
start = time.time()
py_list = [i*2 for i in range(100000)]
print("\n List operation time : ", time.time()-start)

start = time.time()
np_array = np.arange(100000)*2
print("\n Numpy operation time : ", time.time()-start)
```

### Vector,Matrix and Tensor

### Array Reshaping

```python
arr = np.arange(12)
print("Original: ",arr) 

Reshape = arr.reshape((3,4))
print("\n Reshaped_array: ", Reshape)


Flattened = Reshape.flatten()
print("Flattened_array: ",Flattened)


Raveled = Reshape.ravel()  # Return  view , insted of copy
print(" \n reval_array: ",Raveled)


# Transpose

Transponse = Reshape.transpose()
print("Transpose_Arrya: ",Transponse)

```
